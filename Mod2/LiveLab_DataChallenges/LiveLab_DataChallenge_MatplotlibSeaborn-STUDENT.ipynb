### Live Lab AND Data Challenge:  Matplotlib & Seaborn

You are given sample data for visitors that visit a webiste per day and signup for an account.  Your job is to create 8 visuals (4 in matplotlib and 4 in seaborn) to visualize the following: 

-- Line chart of Visitors over Time

-- Histogram of Signups

-- Scatter plot of Visitors vs Signups

-- Bar chart of Average Signups per Visitor Range

Replace the word `None` with your code

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create the dataset
data = {
    'Day': list(range(1, 11)),
    'Visitors': [100, 120, 110, 130, 150, 140, 160, 170, 180, 175],
    'Signups': [5, 8, 6, 10, 12, 11, 13, 15, 16, 14]
}

df = pd.DataFrame(data)

# Optional: Set Seaborn style
sns.set(style="whitegrid")

# Create a figure with 4 rows and 2 columns to make 8 visuals
fig, axs = plt.subplots(4, 2, figsize=(12, 16))
fig.suptitle("Matplotlib vs Seaborn Visualizations", fontsize=16)




# Task 1: Line chart of Visitors over Time -- Matplotlib
axs[0, 0].plot(df['Day'], df['Visitors'], marker='o', color='blue')
axs[0, 0].set_title("Matplotlib: Visitors Over Time")
axs[0, 0].set_xlabel("Day")
axs[0, 0].set_ylabel("Visitors")

# Task 1: Line chart of Visitors over Time -- Seaborn
sns.lineplot(x='Day', y='Visitors', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Seaborn: Visitors Over Time')
axs[0, 1].set_xlabel("Day")
axs[0, 1].set_ylabel("Visitors")



# Task 2: Histogram of Signups -- MatplotLib
axs[1, 0].hist(df['Signups'], bins=5, color='orange', edgecolor='black')
axs[1, 0].set_title("Matplotlib: Signup Distribution")
axs[1, 0].set_xlabel("Signups")
axs[1, 0].set_ylabel("Frequency")

# Task 2: Histogram of Signups -- Seaborn
sns.histplot(data=df, x= 'Signups', bins=5, kde=False, color='Orange', ax=axs[1, 1])
axs[1, 1].set_title("Seaborn: Signup Distribution")
axs[1, 1].set_xlabel('Signups')
axs[1, 1].set_ylabel("Frequency")




# Task 3: Scatter plot of Visitors vs Signups -- Matplotlib
axs[2, 0].scatter(df['Signups'], df['Visitors'], color='green')
axs[2, 0].set_title("Matplotlib: Visitors vs Signups")
axs[2, 0].set_xlabel("Visitors")
axs[2, 0].set_ylabel("Signups")

# Task 3: Scatter plot of Visitors vs Signups -- Seaborn
sns.scatterplot(x='Signups', y='Visitors', data=df, ax=axs[2,1], color='Green')
axs[2, 1].set_title("Seaborn: Visitors vs Signups")
axs[2, 1].set_xlabel("Visitors")
axs[2, 1].set_ylabel('Signups')



# Creating a variable for a bar chart -- what does this code do?
df['Visitor_Bin'] = pd.cut(df['Visitors'], bins=3)
signup_avg = df.groupby('Visitor_Bin')['Signups'].mean().reset_index()




# Task 4: Bar chart of average Signups per Visitor Bin - Matplotlib
axs[3, 0].bar(signup_avg['Visitor_Bin'].astype(str), signup_avg['Signups'], color='purple')
axs[3, 0].set_title("Matplotlib: Avg Signups by Visitor Range")
axs[3, 0].set_xlabel('Visitor Range')
axs[3, 0].set_ylabel('Average Signups')


# Task 4: Bar chart of average Signups per Visitor Bin - Seaborn
sns.barplot(data=signup_avg, y='Signups', x= "Visitor_Bin", ax=axs[3, 1], palette='muted')
axs[3,1].set_title('Seaborn: Average signups per visitor bin')
axs[3,1].set_xlabel("Visitor Range")
axs[3,1].set_ylabel("Average Signups")




# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

- Answer: Yes, the number of app visitors is generally increasing over time, with some minor drops on certain days. The overall trend shows growth in visitor activity.

- Question: Do more visitors lead to more signups?
- Answer: Based off the graph, yes as the numvers of visitors increases, the number of sign ups gradally increase too.
