```python
import numpy as np, pandas as pd
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X, _= make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=0)
X_df = pd.DataFrame(X, columns=['feature1', 'feature2'])
print( 'X_df shape:', X_df.shape)
Z = StandardScaler().fit_transform(X_df)
Ks = range (2, 11)
sil, ch, db = [], [], []
for k in Ks:
    km = KMeans(n_clusters=k, n_init=10, random_state=0).fit(Z)
    labels = km. labels_
    sil.append (silhouette_score(Z, labels))
    ch.append (calinski_harabasz_score(Z, labels))
    db.append (davies_bouldin_score(Z, labels))
print(list(zip(Ks, np. round(sil,3), np. round(ch,1), np. round (db,3) )) [:5])
```

```python
from scipy.spatial.distance import cdist
import numpy as np
K_FINAL = 4
km = KMeans(n_clusters = K_FINAL, n_init = 10, random_state = 42 ).fit(Z)
labels = km. labels_
centers = km.cluster_centers_
# within spread (avg feature std per cluster)
within = [Z[labels==c].std(axis=0).mean() for c in range(K_FINAL)]
max_within = float (np.max(within))
# between-centroid distances
D = cdist(centers, centers)
np. fill_diagonal (D, np.nan)
min_between = float (np. nanmin (D))
print(f"Separation ratio = {min_between/max_within:.2f} (higher is better)")
```
