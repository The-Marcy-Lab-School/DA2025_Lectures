```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
dates = pd.date_range('2021-01-01', periods=48, freq='M')
data = np.arange(1,49) + 10*np.sin(np.arange(1,49)*(2*np.pi/12)) + np.random.randn(48)*2
ts = pd.Series(data, index=dates, name='Sales')
decomp = seasonal_decompose(ts, model='additive', period=12)
fig = decomp.plot(); fig.set_size_inches(10,8); plt.show()
```
